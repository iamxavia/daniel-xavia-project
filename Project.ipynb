```python
import pandas as pd
import plotly.express as px
#read data
df = pd.read_csv("/content/drive/MyDrive/midterm.csv", low_memory=False)
```

```python
#clean data
def clean_data():
  df.isnull()
  df.dropna(inplace=True)
  return df
clean_data()

#tinh tong
tong_diem=[]
for i in df.index:
  grade = df.loc[i][3:]
  tong_diem.append(grade.sum())
df['Sum']=tong_diem 
df

#tong_phan_3
tong_diem_3=[]
for i in df.index:
  grade_3 = df.loc[i][5:9]
  tong_diem_3.append(grade_3.sum())
df['Tong phan 3']=tong_diem_3 
df
```
